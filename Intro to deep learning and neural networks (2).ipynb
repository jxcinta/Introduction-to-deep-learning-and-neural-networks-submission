PART A

```python
import pandas as pd
import numpy as np
```

Looking at the data

```python
concrete_data = pd.read_csv('concrete_data.csv') 
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

Seperating target data and what data is the predictors

```python
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] 
target = concrete_data['Strength'] 
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1] 
print(n_cols) #checking everything is correct
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

Creating the model with one hidden layer with 10 nodes

```python
def regression_model():

    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

Testing and training the model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
model = regression_model()
epochs = 50
model.fit(X_train, y_train, epochs=epochs, verbose=1)
```

```python
loss = model.evaluate(X_test, y_test)
print("The loss value is ", loss)
y_pred = model.predict(X_test)
```

Calculating the mean square error

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
mean = np.mean(mse)
standard_deviation = np.std(mse)
print("The mean is ", mean, "The standard deviation is ", standard_deviation)
```

Creating the loop

```python
runs = 50
epochs = 50
mse_list = []

for i in range(0, runs):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=i)
    model.fit(X_train, y_train, epochs=epochs, verbose=0)
    meansqrerror = model.evaluate(X_test, y_test, verbose=0)
    print(meansqrerror)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)
```

```python
mse_list = np.array(mse_list)
mean = np.mean(mse_list)
standard_deviation = np.std(mse_list)

print("the mean is ", mean)
print("The standard devation is", standard_deviation)
```

PART D

3 hidden layers each with 10 nodes using the relu activation function

```python
#Part D
def regression_model():

    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
model.fit(predictors_norm, target, validation_split=0.3, epochs=100, verbose=2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
model = regression_model()
epochs = 50
model.fit(X_train, y_train, epochs=epochs, verbose=1)
```

```python
loss = model.evaluate(X_test, y_test)
print("Loss value is", loss)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mean = np.mean(mse)
standard_deviation = np.std(mse)
print("The mean is ", mean, "The standard deviation is ", standard_deviation)
```

```python
runs = 50
epochs = 50
mse_list = []

for i in range(0, runs):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=i)
    model.fit(X_train, y_train, epochs=epochs, verbose=0)
    meansqrerror = model.evaluate(X_test, y_test, verbose=0)
    print(meansqrerror)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_list.append(mse)
```
